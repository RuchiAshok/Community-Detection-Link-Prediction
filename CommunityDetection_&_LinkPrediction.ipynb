<h2>2. Louvain Method </h2>

This method is divided in 2 phases: Modularity Optimization and Community Aggregation.

```python
def simple_Louvain(G):
    partition = community.community_louvain.best_partition(G)
    
    max_k_w = []
    for com in set(partition.values()):
        list_nodes = [nodes for nodes in partition.keys()
                      if partition[nodes] == com]
        max_k_w = max_k_w + [list_nodes]

    
    node_mapping = {}
    map_v = 0
    for node in G.nodes():
        node_mapping[node] = map_v
        map_v += 1

    community_num_group = len(max_k_w)
    color_list_community = [[] for i in range(len(G.nodes()))]
    
    # color
    for i in G.nodes():
        for j in range(community_num_group):
            if i in max_k_w[j]:
                color_list_community[node_mapping[i]] = j
    
    return  max_k_w
```

```python
lo_comm_groups = simple_Louvain(G)
print("Number of communities detected: ",len(lo_comm_groups))
```

```python
nodes_in_communities(lo_comm_groups)
```

```python
community_visualization(lo_comm_groups)
```

<h2>3. Surprise Communities</h3>

<p>The algorithm is almost similar to the Louvain community detection algorithm except that it uses surprises instead of modularity. </p>
<p>Nodes are moved from one community to another such that surprises are greedily improved. This approach considers the probability that a link lies within a community.</p>

<p>The use of surprises works well in the limit of many small communities and the use of modularity works well in the limit of a few large communities.</p>

<h1><b>Link Prediction </h1>

<b> Adjacency Matrix </b> 

```python
#Storing the graph's adjacency matrix in a numpy array
A = nx.adjacency_matrix(G).todense()
len(A[0])
A
```

Storing degree of each nodes in a dictionary

```python
degree_of_nodes={}
#node index : degree value
x=A.sum(axis=1).tolist()
for i in range(len(x)):
  degree_of_nodes[i]=x[i][0]
```

```python
degree_of_nodes
```

<h2>1. PAC</h2>

<p>For each pair of non connected node pair (Ni,Nj), PAC score is calculated by</p><p>
<b>PAC Score of (Ni,Nj) = degree of (Ni)* degree of (Nj)</p>

```python
'''
For each node,  calculated PAC score with all other nodes and store it in a list as
node_pair_pac_score - list [node1, node2, PAC_Score]

'''
# node_pair_pac_score - list [node1, node2, PAC_Score]
def PAC_each_node(node,index):
  node_pair_pac_score =[] 
  for n in range(index,len(degree_of_nodes)):
    p =[]
    if(n!= node):
      p.append(node) #node1
      p.append(n) #node2
      value=degree_of_nodes[node]*degree_of_nodes[n]
      p.append(value)   #PAC Score
      node_pair_pac_score.append(p)
  return node_pair_pac_score
```

```python
#Find PAC Scores for all the possible combinations of node pairs
for n in range(len(degree_of_nodes)):
    node_pair_pac_matrix =[]
    #node_pair_pac_matrix : PAC Score of 561 (34 C 2 possible combination)
    for node in range(len(degree_of_nodes)+1):
      index = node
      node_pair_pac = []
      node_pair_pac = PAC_each_node(node,index)
      node_pair_pac_matrix.append(node_pair_pac)
```

```python
#Storing all possible combinations of node pairs as list
pac_score_final_list = []
for nodepair in range(len(node_pair_pac_matrix)):
  for each in range(len(node_pair_pac_matrix[nodepair])):
    pac_score_final_list.append(node_pair_pac_matrix[nodepair][each])

print("Total node pair commbination possible: ", len(pac_score_final_list) )  #34C2 = 561
```

```python
PAC_List_NodePair_Desc=sorted(pac_score_final_list, key = lambda x:x[2], reverse = True) 
PAC_edges_new = [] #Contains Top 5 Link Predicted node pairs
print("Top 5 node pair link prediction based on PAC Score are:")
for i in range(5):
  print(PAC_List_NodePair_Desc[i][0]," : ",PAC_List_NodePair_Desc[i][1]," with PAC Score: ",PAC_List_NodePair_Desc[i][2])
  PAC_edges_new.append((PAC_List_NodePair_Desc[i][0],PAC_List_NodePair_Desc[i][1]))
```
